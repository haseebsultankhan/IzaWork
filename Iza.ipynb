```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
from IPython import display
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel(r'C:/Users/user1/Desktop/Iza/vpa.xlsx')
print(df)
```

# CREATING A MODEL TO CHECK PVA between DW(dependent variable) with Precipitates+Particulate+biological (Independent Variables) 

```python
mod = ols('DW~Precipitates+Particulate+biological' , data= df).fit()  
```

# APPLYING STATS (I'm not sure about type, we can use type = 1 or type = 2 or type = 3) , it doesn't make any difference in the results

## FOR UNDERSTANDING WHAT IS TYPE , see this reference link
## https://mcfromnz.wordpress.com/2011/03/02/anova-type-iiiiii-ss-explained/

```python
aov = sm.stats.anova_lm(mod, type=1)
```

```python
print(aov)
```

```python
aov2 = sm.stats.anova_lm(mod, type=2)
```

```python
print(aov2)
```
